```python
# DATA EXPLORATION

import os
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
rootdir = 'Eardata/Images'
folders = []
count=0
for file in os.listdir(rootdir):
    folders.append(file)
    count=count+1
```

```python
# TOTAL NO. OF FOLDERS

count
```

```python
# NAMES OF ALL FOLDERS IN DATASET

folders
```

```python

x=[]
y=[]
no_of_images= []

for dire in folders:
    path = rootdir + '/'+dire
    count=0
    for image in os.listdir(path):
        image_path = path + '/' + image
        img = Image.open(image_path)
        count=count+1
        width,height = img.size
        x.append(height)
        y.append(width)
    no_of_images.append(count)
        
```

```python
#  NO. OF IMAGES IN EACH FOLDER
no_of_images
```

```python
# TOTAL NO. OF IMAGES

len(x)
```

```python
len(y)
```

```python
# HEIGHTS OF ALL IMAGES
x
```

```python
# WIDTH OF ALL IMAGES
y
```

```python
ht = dict()
for i in range(len(x)):
    if x[i] in ht.keys(): 
        ht[x[i]] += 1
    else: 
        ht[x[i]] = 1
print("HEIGHT  FREQUENCY")
for u in ht: 
    print(u, "     ", ht[u]) 
```

```python
wt = dict()
for i in range(len(y)):
    if y[i] in wt.keys(): 
        wt[y[i]] += 1
    else: 
        wt[y[i]] = 1
print("WIDTH  FREQUENCY")      
for u in wt: 
    print(u, "     ", wt[u]) 
```

```python
sno= []
count=1
for i in range(len(folders)):
    sno.append(count)
    count=count+1
```

```python
# NO. OF SAMPLES PER PERSON

plt.figure(figsize=(15, 8)) 
plt.bar(sno,no_of_images,width = 0.3, color = ['red'])
plt.title('No. of Images in each folder')
plt.xlabel('Folders')
plt.ylabel('No. of Images')
plt.show()
```

```python
plt.figure(figsize=(12, 5)) 
plt.bar(*zip(*ht.items()),width = 0.8)
plt.title('Height of Images V/S No. of images corresponding a height')
plt.xlabel('Height(in pixels)')
plt.ylabel('No. of Images')
plt.show()
```

```python
plt.figure(figsize=(12, 5)) 
plt.bar(*zip(*wt.items()),width = 0.6)
plt.title('Width of Images V/S No. of images corresponding a width')
plt.xlabel('Width(in pixels)')
plt.ylabel('No. of Images')
plt.show()
```

```python
# MAXIMUM HEIGHT CORRESONDING TO EACH FOLDER OR
# MAXIMUM HEIGHT OF IMAGE CORRESPONDING TO EACH PERSON


i=0
max_height_all=[]
for c in no_of_images:
    height_list=[]
    for h in range(c):
        height_list.append(x[i+h])
    maximum_height=max(height_list)
    max_height_all.append(maximum_height)
    i=i+c
```

```python
# MAXIMUM WIDTH CORRESONDING TO EACH FOLDER OR
# MAXIMUM WIDTH OF IMAGE CORRESPONDING TO EACH PERSON

i=0
max_width_all=[]
for c in no_of_images:
    width_list=[]
    for h in range(c):
        width_list.append(y[i+h])
    maximum_width=max(width_list)
    max_width_all.append(maximum_width)
    i=i+c
```

```python
# MINIMUM WIDTH CORRESONDING TO EACH FOLDER OR
# MINIMUM WIDTH OF IMAGE CORRESPONDING TO EACH PERSON

i=0
min_width_all=[]
for c in no_of_images:
    width_list=[]
    for h in range(c):
        width_list.append(y[i+h])
    minimum_width=min(width_list)
    min_width_all.append(minimum_width)
    i=i+c
```
